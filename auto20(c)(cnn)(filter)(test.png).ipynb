```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('test.png',cv2.IMREAD_GRAYSCALE)
```

```python
image
```

```python
filter1 = np.array([[0,1,0],[1,-4,1],[0,1,0]])
filter1
```

```python
filtered_image = cv2.filter2D(image,-1,filter1)
```

```python
filtered_image
```

```python
print('filtered image pixels')
for i in range(filtered_image.shape[0]):
    for j in range(filtered_image.shape[1]):
        print(filtered_image[i,j],end=' ')
    print()
```
